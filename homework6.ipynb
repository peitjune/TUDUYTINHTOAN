### W6A16

Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python
n,m=map(int,input().split())
a=[list(map(int,input().split())) for _ in range(n)]
for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:>4}",end="")
    print()
```

### W6A17

Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và đường chéo phụ của mảng

```python
n=int(input())
a=[list(map(int,input().split())) for _ in range(n)]
print(*[a[i][i] for i in range(n)])
print(*[a[i][n-1-i] for i in range(n)])
```

### W6A19

Viết hàm tìm vị trí giá trị lớn nhất

```python
def max_index(lst):
    idx=0
    for i in range(len(lst)):
        if lst[i]>lst[idx]:
            idx=i
    return idx
```

### W6A20

Tìm vị trí đầu tiên của k

```python
def find_first(lst,k):
    for i in range(len(lst)):
        if lst[i]==k:
            return i
    return -1
```

### W6A21

Tạo dict ánh xạ chuỗi -> tuple vị trí

```python
words=input().split()
d={}
for i in range(len(words)):
    if words[i] not in d:
        d[words[i]]=[]
    d[words[i]].append(i)
for k in d:
    d[k]=tuple(d[k])
print(d)
```

```python
def remove_less(lst,x):
    return [i for i in lst if i>=x]
```

### W6A24

Đẩy số 0 về đầu

```python
def move_zero(lst):
    return [0 for i in lst if i==0]+[i for i in lst if i!=0]
```

### W6A25

Tìm số thiếu 0-9

```python
def missing(lst):
    return [i for i in range(10) if i not in lst]
```

### W6A26

Tổng cực đại địa phương

```python
def local_max_sum(lst):
    s=0
    for i in range(1,len(lst)-1):
        if lst[i]>lst[i-1] and lst[i]>lst[i+1]:
            s+=lst[i]
    return s
```

### W6A28

Kiểm tra tăng/giảm

```python
lst=list(map(int,input().split()))
tang=all(lst[i]>lst[i-1] for i in range(1,len(lst)))
giam=all(lst[i]<lst[i-1] for i in range(1,len(lst)))
print("Tang" if tang else "Giam" if giam else "Vo danh")
```
