```python
lst = [1, 2, 3, "toi", 5.7]
print(lst) # in ra list
print(*lst) # * nghĩa là unpack, không kèm [] và , và """
for x in lst: # duyệt lst từng giá trị
    print(x, end = " ") # như print(*lst)

print(lst[0]) # in phần tử đầu tiên
print(lst[len(lst) - 1]) # in phần tử cuối
print(lst[-1]) # như lệnh trên
```

```python
lst = [1, 2, 3, "toi", 5.7]
for i, x in enumerate(lst): # duyệt list lần lượt
    print(f"lst[{i}] = {x}")
for i, x in enumerate(lst, 2): # duyệt list lần lượt, nhưng cho giá trị "1" với stt 0 thành stt 2 và đếm tiếp
    print(f"lst[{i}] = {x}")
```

```python
lst = [1, 2, 3, "toi", 5.7]
lst.append([1, 2, 5]) # append là thêm phần tử, ở đây thêm ptu [1, 2, 5] (đây là 1 ptu)
lst.append(12) # thêm số 12
print(lst)
```

```python
lst = [1, 2, 3, "toi", 5.7]
lst.insert(0, 100) # insert cũng là thêm phần tử nhưng có thể ghi vị trí cần thêm, ở đây thêm số 100 vào vtri 0
print(lst[-1]) # in gtri ở vị trí -3, dạng [0, 1, 2, 3, 4] nếu là dương, [-5, -4, -3, -2, -1] nếu âm 
```

```python
lst = [1, 2, 3, "toi", 5.7]
lst = lst + [1, 2] # thêm list
print(lst)
```

```python
lst = [1, 2, 3, "toi", 5.7]
lst.remove(1) # xóa gtri 1
print(lst)
```
