```python
s = input()
number = 0
tong = 0
count = 0
for c in s:
    if "0" <= c <= "9":
        number = number*10 + (ord(c) - ord("0"))
    else:
        tong += number
        number = 0
tong += number
sum = (tong/count)
print(sum)
```

```python
s = input()
sum = 0
count = 0
for i in s:
    if i.isdigit():
        count += 1
    elif i.isalpha():
        count += 0
for i in s:
    if i.isdigit():
        i = int(i)
        sum += i
    
tong = (sum/count)
print(f"{tong:.2f}")
```

```python
def prime(n):
    if (n < 2):
        return False
    for i in range(2, int(n**0.5)+1):
        if (n % i == 0):
            return False
    return True

while True:
    try:
        n = int(input())
        if prime(n) == True:
            print("Day la so nguyen to")
            break
        else:
            print("Day khong phai")
            break
    except:
        print("Nhap lai")
```
