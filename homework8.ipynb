**W8A16:** Yêu cầu: Xây dựng lớp `Shape` với phương thức `area()`. Xây dựng lớp `Rectangle(w, h)` và `Circle(r)` kế thừa từ `Shape`, ghi đè `area()`. Lấy $\Pi=3.14$.

```python
class Shape:
    def area(self): pass

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = float(w)
        self.h = float(h)
    def area(self):
        return self.w * self.h

class Circle(Shape):
    def __init__(self, r):
        self.r = float(r)
    def area(self):
        return 3.14 * (self.r ** 2)
```

**W8A17:** Yêu cầu: Xây dựng lớp `Vehicle(brand, speed)`. Xây dựng lớp `Car(brand, speed, seats)` kế thừa từ `Vehicle`. Viết phương thức `info()` cho mỗi lớp.

```python
class Vehicle:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed
    def info(self):
        return f"{self.brand}-{self.speed}"

class Car(Vehicle):
    def __init__(self, brand, speed, seats):
        super().__init__(brand, speed)
        self.seats = seats
    def info(self):
        return f"{self.brand}-{self.speed}-{self.seats}"
```

**W8A18:** Yêu cầu: Xây dựng lớp `Polygon` với phương thức `perimeter()`. Xây dựng lớp `Triangle(a, b, c)` kế thừa từ `Polygon` và ghi đè `perimeter()`.

```python
class Polygon:
    def perimeter(self): pass

class Triangle(Polygon):
    def __init__(self, a, b, c):
        self.a = int(a)
        self.b = int(b)
        self.c = int(c)
    def perimeter(self):
        return self.a + self.b + self.c
```

**W8A19:** Yêu cầu: Xây dựng lớp `Person(name, age)` và lớp `Student(name, age, school)` kế thừa từ `Person`. Phương thức `describe()`.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def describe(self):
        return f"{self.name} {self.age}"

class Student(Person):
    def __init__(self, name, age, school):
        super().__init__(name, age)
        self.school = school
    def describe(self):
        return f"{self.name} {self.age} {self.school}"
```

**W8A20:** Yêu cầu: Xây dựng lớp `Account(balance)` với hai phương thức `deposit(x)` và `get_balance()`. Xây dựng lớp `SavingAccount(balance, rate)` kế thừa từ `Account`.

```python
class Account:
    def __init__(self, balance):
        self.balance = float(balance)
    def deposit(self, x):
        self.balance += float(x)
    def get_balance(self):
        return self.balance

class SavingAccount(Account):
    def __init__(self, balance, rate):
        super().__init__(balance)
        self.rate = float(rate)
    def add_interest(self):
        self.balance += self.balance * self.rate
```

**W8A21:** Yêu cầu: Xây dựng lớp `Printer`. Xây dựng lớp `BinaryPrinter` kế thừa từ `Printer`, ghi đè `print_value(x)` để in biểu diễn nhị phân.

```python
class Printer:
    def print_value(self, x): pass

class BinaryPrinter(Printer):
    def print_value(self, x):
        print(bin(int(x))[2:])
```

**W8A22:** Yêu cầu: Xây dựng lớp `Matrix2x2(a,b,c,d)`. Viết phương thức `trace()` và `determinant()`.

```python
class Matrix2x2:
    def __init__(self, a, b, c, d):
        self.a, self.b, self.c, self.d = int(a), int(b), int(c), int(d)
    def trace(self):
        return self.a + self.d
    def determinant(self):
        return self.a * self.d - self.b * self.c
```

**W8A23:** Yêu cầu: Xây dựng lớp `Polynomial` biểu diễn đa thức bậc nhất ax + b. Viết phương thức `evaluate(x)`.

```python
class Polynomial:
    def __init__(self, a, b):
        self.a, self.b = float(a), float(b)
    def evaluate(self, x):
        return self.a * float(x) + self.b
```

**W8A24:** Yêu cầu: Xây dựng lớp `Polynomial2` biểu diễn đa thức bậc hai ax^2+bx+c. Viết phương thức `evaluate(x)`.

```python
class Polynomial2:
    def __init__(self, a, b, c):
        self.a, self.b, self.c = float(a), float(b), float(c)
    def evaluate(self, x):
        return self.a * (float(x) ** 2) + self.b * float(x) + self.c
```

**W8A25:** Yêu cầu: Xây dựng lớp `PolynomialSum` cho hai đa thức bậc nhất. Viết phương thức `sum_str()` trả về chuỗi biểu diễn đa thức tổng.

```python
class PolynomialSum:
    def __init__(self, a, b, c, d):
        self.k = float(a) + float(c)
        self.m = float(b) + float(d)
    def sum_str(self):
        k_str = int(self.k) if self.k.is_integer() else self.k
        m_str = int(self.m) if self.m.is_integer() else self.m
        return f"{k_str}x + {m_str}"
```

**W8A27:** Yêu cầu: Xây dựng lớp `Product(name, price)` và lớp `DiscountProduct` ghi đè phương thức `final_price()`.

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = float(price)

class DiscountProduct(Product):
    def __init__(self, name, price, percent):
        super().__init__(name, price)
        self.percent = float(percent)
    def final_price(self):
        return self.price * (100 - self.percent) / 100
```

**W8A28:** Yêu cầu: Xây dựng lớp `Counter` có thuộc tính count ban đầu bằng 0 và các phương thức `inc()`, `dec()`, `value()`.

```python
class Counter:
    def __init__(self):
        self.count = 0
    def inc(self):
        self.count += 1
    def dec(self):
        self.count -= 1
    def value(self):
        return self.count
```

**W8A29:** Yêu cầu: Xây dựng lớp `Time(h, m, s)` với phương thức `to_seconds()`.

```python
class Time:
    def __init__(self, h, m, s):
        self.h, self.m, self.s = int(h), int(m), int(s)
    def to_seconds(self):
        return self.h * 3600 + self.m * 60 + self.s
```

**W8A30:** Yêu cầu: Xây dựng lớp `Stack` sử dụng danh sách để cài đặt ngăn xếp với các phương thức `push(x)`, `pop()`, `top()`.

```python
class Stack:
    def __init__(self):
        self.items = []
    def push(self, x):
        self.items.append(x)
    def pop(self):
        if self.items: self.items.pop()
    def top(self):
        return self.items[-1] if self.items else None
```
