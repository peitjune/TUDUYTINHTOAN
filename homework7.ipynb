**W7A16:** Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu mỗi phần tử x trong mảng xuất hiện đúng x lần. Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp.

```python
def make_beautiful_array(arr):
    counts = {}
    for num in arr:
        counts[num] = counts.get(num, 0) + 1
    removals = 0
    for x, count in counts.items():
        if count < x: removals += count
        elif count > x: removals += (count - x)
    return removals
```

**W7A17:** Tìm phần tử xuất hiện nhiều hơn $n/2$ lần trong mảng (nếu có) với độ phức tạp thời gian là $O(n)$, độ phức tạp bộ nhớ là $O(1)$, - Boyer-Moore. Nếu không có phần tử nào thỏa mãn thì không in gì.

```python
def majority_element(arr):
    candidate = None
    count = 0
    for num in arr:
        if count == 0:
            candidate = num
            count = 1
        elif num == candidate:
            count += 1
        else:
            count -= 1
    if arr.count(candidate) > len(arr) // 2:
        print(candidate)
```

**W7A18:** Sắp xếp có điều kiện: Số chẵn đứng trước, số lẻ đứng sau, chẵn tăng dần, lẻ giảm dần.

```python
def custom_sort(arr):
    evens = [x for x in arr if x % 2 == 0]
    odds = [x for x in arr if x % 2 != 0]
    for i in range(len(evens)):
        for j in range(i + 1, len(evens)):
            if evens[i] > evens[j]: evens[i], evens[j] = evens[j], evens[i]
    for i in range(len(odds)):
        for j in range(i + 1, len(odds)):
            if odds[i] < odds[j]: odds[i], odds[j] = odds[j], odds[i]
    return evens + odds
```

**W7A19:** Merge 2 mảng đã sắp xếp (có thể tăng dần, có thể giảm dần) mà không dùng bất cứ thuật toán sắp xếp nào.

```python
def merge_sorted_arrays(arr1, arr2):
    if len(arr1) > 1 and arr1[0] > arr1[-1]: arr1 = arr1[::-1]
    if len(arr2) > 1 and arr2[0] > arr2[-1]: arr2 = arr2[::-1]
    res = []
    i, j = 0, 0
    while i < len(arr1) and j < len(arr2):
        if arr1[i] < arr2[j]:
            res.append(arr1[i])
            i += 1
        else:
            res.append(arr2[j])
            j += 1
    res.extend(arr1[i:])
    res.extend(arr2[j:])
    return res
```

**W7A22:** Cho N tên sinh viên gồm cả họ và tên, tìm xem tên nào hay gặp nhất. Nếu có nhiều tên cùng số lần gặp thì lấy tên có thứ tự từ điển nhỏ nhất.

```python
def most_frequent_name(names):
    counts = {}
    for name in names:
        counts[name] = counts.get(name, 0) + 1
    max_freq = max(counts.values())
    candidates = [name for name, count in counts.items() if count == max_freq]
    best = candidates[0]
    for name in candidates[1:]:
        if name < best: best = name
    return best
```

**W7A23:** Đếm số lượng tên trong xâu kí tự. Tên được định nghĩa là xâu có độ dài >=2 và có chữ cái đầu viết hoa, các chữ cái khác viết thường, chỉ có kí tự cái.

```python
def count_valid_names(s):
    words = s.split()
    count = 0
    for word in words:
        if len(word) >= 2 and word.isalpha() and word[0].isupper() and word[1:].islower():
            count += 1
    return count
```

**W7A24:** Cho xâu kí tự, xóa các từ giống nhau liền kề, chỉ giữ lại từ đầu tiên.

```python
def remove_adjacent_duplicates(s):
    words = s.split()
    if not words: return ""
    result = [words[0]]
    for word in words[1:]:
        if word != result[-1]: result.append(word)
    return " ".join(result)
```

**W7A25:** Tìm số có đúng k chữ số lớn nhất có thể trong một số có n chữ số (coi đầu vào là một số rất lớn - str). Không được hoán đổi vị trí các chữ số.

```python
def largest_k_digit(s, k):
    drop = len(s) - k
    stack = []
    for digit in s:
        while drop > 0 and stack and stack[-1] < digit:
            stack.pop()
            drop -= 1
        stack.append(digit)
    return "".join(stack[:k])
```

**W7A26:** Tách các số có trong một đoạn văn bản, tính tổng và tính trung bình các số đó.

```python
def extract_and_calculate(text):
    nums = []
    curr = ""
    for char in text:
        if char.isdigit(): curr += char
        else:
            if curr:
                nums.append(int(curr))
                curr = ""
    if curr: nums.append(int(curr))
    if not nums: return 0, 0
    total = sum(nums)
    avg = total / len(nums)
    return total, avg
```

**W7A27:** Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra xem dãy ngoặc đã cho có hợp lệ hay không.

```python
def is_valid_parentheses(s):
    stack = []
    pairs = {')': '(', '}': '{', ']': '['}
    for char in s:
        if char in "({[": stack.append(char)
        elif char in ")}]":
            if not stack or stack[-1] != pairs[char]: return False
            stack.pop()
    return len(stack) == 0
```

**W7A28:** Bài toán Tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên nào <=N (N là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ dài thì lấy dãy nhỏ nhất.

```python
def is_prime(num):
    if num < 2: return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0: return False
    return True

def longest_composite_sequence(arr, N):
    max_seq = []
    curr_seq = []
    for num in arr:
        if num > N and not is_prime(num):
            curr_seq.append(num)
        else:
            if len(curr_seq) > len(max_seq): max_seq = curr_seq
            curr_seq = []
    if len(curr_seq) > len(max_seq): max_seq = curr_seq
    return max_seq
```

**W7A29:** Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau.

```python
def lcis(arr):
    if not arr: return []
    max_seq = [arr[0]]
    curr_seq = [arr[0]]
    for i in range(1, len(arr)):
        if arr[i] > arr[i-1]:
            curr_seq.append(arr[i])
        else:
            if len(curr_seq) > len(max_seq): max_seq = curr_seq
            curr_seq = [arr[i]]
    if len(curr_seq) > len(max_seq): max_seq = curr_seq
    return max_seq
```

**W7A30:** Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập vào từ bàn phím.

```python
def lis(arr):
    if not arr: return []
    dp = [[x] for x in arr]
    for i in range(1, len(arr)):
        for j in range(i):
            if arr[i] > arr[j] and len(dp[j]) + 1 > len(dp[i]):
                dp[i] = dp[j] + [arr[i]]
    longest = []
    for seq in dp:
        if len(seq) > len(longest): longest = seq
    return longest
```
