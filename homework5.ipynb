### W5A16

Viết hàm kiểm tra một số có phải số Armstrong hay không. Hàm trả về bool.

```python
def is_armstrong(x):
    num_str = str(x)
    n = len(num_str)
    total = sum(int(digit) ** n for digit in num_str)
    return total == x
number = int(input())
is_armstrong(number)
```

### W5A19
VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên. 

```python
def is_magic(x):
    s=sum(int(d) for d in str(x))
    while s>9: s=sum(int(d) for d in str(s))
    return s==1
```

### W5A20

Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'

```python
def detect_number_type(s):
    try: int(s); return 'int'
    except:
        try: float(s); return 'float'
        except: return 'neither'
```

### W5A25

Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay 
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3), 
(0,3)], p = (2,1) 

```python
def point_in_quadrilateral(quad,p):
    xs=[pt[0] for pt in quad]; ys=[pt[1] for pt in quad]
    return min(xs)<=p[0]<=max(xs) and min(ys)<=p[1]<=max(ys)
```

### W5A28

Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3). 

```python
import math
def angle_ABC(A,B,C):
    BA=(A[0]-B[0],A[1]-B[1]); BC=(C[0]-B[0],C[1]-B[1])
    dot=BA[0]*BC[0]+BA[1]*BC[1]
    norm=math.sqrt(BA[0]**2+BA[1]**2)*math.sqrt(BC[0]**2+BC[1]**2)
    return math.degrees(math.acos(dot/norm))
```

### W5A34

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau 
Nếu n ≤ 0 trả về 0.

```python
def sum_1_to_n(n):
    if n<=0: return 0
    return sum(x for x in range(1,n+1) if len(set(str(x)))==len(str(x)))
```
