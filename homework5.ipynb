kiểm tra số chẵn

```python
## nhap so
n = int(input())
## tạo 1 hàm ktra số chẵn, hàm trả về là true false(bool)
def even(n):
    return (n%2 == 0)
## điều kiện nếu n là số chẵn sử dụng hàm def even:
if(even(n)):
    print(f"{n} la so chan")
else:
    print(f"{n} khong phai la so chan")

```

Kiểm tra số nguyên tố

2 10

```python

def primeNumber(n):
    if (n < 2):
        return False
    for i in range(2, int(n**0.5) + 1):
        if (n % 1 == 0):
            return False
count = 0
a, b = map(int, input().split())
for x in range(a, b + 1):
    if(primeNumber(x) == True):
        count += x
print(count)
```

Tính và so sánh 2 BMI

```python
def compareBMI(w1, h1, w2, h2):
    bmi1 = w1/(h1 * h1)
    bmi2 = w2/(h2 * h2)
    if(bmi1 > bmi2):
        return 1
    elif(bmi1 < bmi2):
        return 2
    else:
        return 0

w1, h1 = map(float, input("Nhap w1 h1").split())
w2, h2 = map(float, input("Nhap w2 h2").split())
if(compareBMI(w1, h1, w2, h2) == 1):
    print("Nguoi thu nhat beo hon nguoi thu hai")
elif(compareBMI(w1, h1, w2, h2) == 2):
    print("Nguoi thu nhat gay hon nguoi thu hai")
else:
    print("Hai nguoi bang nhau")
```

cho 1 điểm tên s, trả lại điểm hệ chữ

```python
def map_grade(s):
    if (s >= 9):
        return "A+"
    elif (s >= 8.5):
        return "A"
    elif (s >= 8):
        return "B+"
    elif (s >= 7):
        return "B"
    else:
        return "Qua ngu"

s = float(input("Nhap diem"))
print(map_grade(s))
```
