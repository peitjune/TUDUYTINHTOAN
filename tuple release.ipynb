```python
tup = (1, 2, 4, 5, 9.0, "rizz") # như list nhưng KHÔNG THÊM BỚT THAY ĐỔI TRONG NÓ, sử dụng ()
lst = list(tup) # lst là list nhưng gtri lấy từ tup
lst[0] = 100 # thay đổi gtri stt 0 thành số 100
tup = tuple(lst) # tup là tuple nhưng lấy gtri từ list
print(tup)
```

```python
tup = (1, 2, 4, 5, 9.0, "rizz")
print(tup[1]) # in giá trị stt 1
print(tup[:2]) # in giá trị theo [:n], in gtri từ stt 0 đến n-1
```
