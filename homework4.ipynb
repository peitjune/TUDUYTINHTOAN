## W4A20 – Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
n = int(input("Nhập số nguyên dương n: "))
temp = n
while temp > 1:
    if temp % 2 != 0:
        break
    temp //= 2
if temp == 1:
    print(f"{n} là số dạng 2^k")
else:
    print(f"{n} không phải số dạng 2^k")
```

## W4A21 – Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input("Nhập số nguyên dương n: "))
tong = 0
temp = n
while temp > 0:
    tong += temp % 10
    temp //= 10
print(f"Tổng các chữ số của {n} là: {tong}")
```

## W4A22 – Nhập vào số nguyên dương n, đếm số chữ số chẵn và số chữ số lẻ

```python
n = int(input("Nhập số nguyên dương n: "))
dem_chan = 0
dem_le = 0
temp = n
while temp > 0:
    chu_so = temp % 10
    if chu_so % 2 == 0:
        dem_chan += 1
    else:
        dem_le += 1
    temp //= 10
print(f"Số chữ số chẵn: {dem_chan}")
print(f"Số chữ số lẻ: {dem_le}")
```

## W4A23 – Nhập n. Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n = int(input("Nhập n: "))
k = 0
S = 0
while True:
    if S + (k + 1) >= n:
        break
    k += 1
    S += k
print(f"k = {k}, S(k) = {S}")
```

## W4A24 – Nhập vào A, tìm n nhỏ nhất sao cho 1 + 1/2 + 1/3 + ... + 1/n > A

```python
A = float(input("Nhập A: "))
tong = 0
n = 0
while tong <= A:
    n += 1
    tong += 1 / n
print(f"n nhỏ nhất = {n}, tổng = {tong:.4f}")
```

## W4A25 – Nhập dãy số nguyên (dừng khi nhập -1), in số lớn nhất và nhỏ nhất

```python
so_lon_nhat = None
so_nho_nhat = None
while True:
    x = int(input("Nhập số (-1 để dừng): "))
    if x == -1:
        break
    if so_lon_nhat is None or x > so_lon_nhat:
        so_lon_nhat = x
    if so_nho_nhat is None or x < so_nho_nhat:
        so_nho_nhat = x
print(f"Số lớn nhất: {so_lon_nhat}")
print(f"Số nhỏ nhất: {so_nho_nhat}")
```

## W4A26 – [Fibonacci1] Nhập vào số nguyên dương A, tìm số Fibonacci lớn nhất không vượt quá A

```python
A = int(input("Nhập số nguyên dương A: "))
a, b = 1, 1
while b <= A:
    a, b = b, a + b
print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {a}")
```

## W4A27 – Đếm số từ trong chuỗi (dùng for trên index, không dùng split(), len())

```python
chuoi = input("Nhập chuỗi: ")
dem_tu = 0
trong_tu = False
for i in range(len(chuoi)):
    if chuoi[i] != ' ' and not trong_tu:
        trong_tu = True
        dem_tu += 1
    elif chuoi[i] == ' ':
        trong_tu = False
print(f"Số từ trong chuỗi: {dem_tu}")
```

## W4A28 – In từ đầu tiên trong chuỗi (dùng for trên index, không dùng split())

```python
chuoi = input("Nhập chuỗi: ")
tu_dau = ""
for i in range(len(chuoi)):
    if chuoi[i] == ' ':
        break
    tu_dau += chuoi[i]
print(f"Từ đầu tiên: {tu_dau}")
```

## W4A29 – Đổi chữ hoa thành thường và thường thành hoa trong chuỗi

```python
chuoi = input("Nhập chuỗi: ")
ket_qua = ""
for ky_tu in chuoi:
    if ky_tu.isupper():
        ket_qua += ky_tu.lower()
    else:
        ket_qua += ky_tu.upper()
print(f"Kết quả: {ket_qua}")
```

## W4A30 – Nhập chuỗi gồm 3 số nguyên cách nhau dấu phẩy, tính tổng 3 số đó

Ví dụ: Nhập `3, 12, 15` → Tổng: `30`

```python
chuoi = input("Nhập 3 số nguyên cách nhau dấu phẩy (vd: 3, 12, 15): ")
so = chuoi.split(",")
tong = int(so[0]) + int(so[1]) + int(so[2])
print(f"Tổng: {tong}")
```

## W4A31 – Đếm số ký tự in hoa, in thường và ký tự số trong chuỗi

```python
chuoi = input("Nhập chuỗi: ")
dem_hoa = 0
dem_thuong = 0
dem_so = 0
for ky_tu in chuoi:
    if ky_tu.isupper():
        dem_hoa += 1
    elif ky_tu.islower():
        dem_thuong += 1
    elif ky_tu.isdigit():
        dem_so += 1
print(f"Số ký tự in hoa: {dem_hoa}")
print(f"Số ký tự in thường: {dem_thuong}")
print(f"Số ký tự số: {dem_so}")
```
