## Bài 4 – In ra hình tam giác cân có độ cao h nhập từ bàn phím

Ví dụ h = 4:
```
   *
  * *
 * * *
*******
```

````python
h = int(input("Nhập độ cao h: "))
for i in range(1, h + 1):
    if i < h:
        spaces = ' ' * (h - i)
        stars = '* ' * i
        print(spaces + stars)
    else:
        print('*' * (2 * h - 1))
````

## Bài 9 – Nhập vào số nguyên dương a, kiểm tra a có phải số chính phương không

Số chính phương là số có căn bậc hai là số nguyên (VD: 1, 4, 9, 16, 25…)

````python
a = int(input("Nhập số nguyên dương a: "))
i = 1
la_chinh_phuong = False
while i * i <= a:
    if i * i == a:
        la_chinh_phuong = True
        break
    i += 1
if la_chinh_phuong:
    print(f"{a} là số chính phương")
else:
    print(f"{a} không phải số chính phương")
````

## Bài 10 – Nhập vào số nguyên dương a, kiểm tra a có phải số hoàn hảo không

## Bài 12 – Nhập n. Cho S(k) = 1 + 2 + ... + k. Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

````python
n = int(input("Nhập số nguyên dương n: "))
k = 0
S = 0
while True:
    if S + (k + 1) >= n:
        break
    k += 1
    S += k
print(f"k = {k}, S(k) = {S}")
````

## Bài 13 – Nhập vào A, tìm n nhỏ nhất sao cho 1 + 1/2 + 1/3 + ... + 1/n > A

````python
A = float(input("Nhập A: "))
tong = 0
n = 0
while tong <= A:
    n += 1
    tong += 1 / n
print(f"n nhỏ nhất = {n}, tổng = {tong:.4f}")
````

## Bài 14 – Nhập dãy số nguyên (dừng khi nhập -1), in số lớn nhất và nhỏ nhất

````python
so_lon_nhat = None
so_nho_nhat = None
while True:
    x = int(input("Nhập số (-1 để dừng): "))
    if x == -1:
        break
    if so_lon_nhat is None or x > so_lon_nhat:
        so_lon_nhat = x
    if so_nho_nhat is None or x < so_nho_nhat:
        so_nho_nhat = x
print(f"Số lớn nhất: {so_lon_nhat}")
print(f"Số nhỏ nhất: {so_nho_nhat}")
````

## Bài 15 – Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số (nhiều cách)

````python
n = int(input("Nhập số nguyên dương n: "))

# Cách 1: Dùng vòng lặp chia cho 10
dem = 0
temp = n
while temp > 0:
    dem += 1
    temp //= 10
print(f"Cách 1 – Số chữ số: {dem}")

# Cách 2: Dùng str để duyệt từng ký tự
dem2 = 0
for c in str(n):
    dem2 += 1
print(f"Cách 2 – Số chữ số: {dem2}")

# Cách 3: So sánh với lũy thừa 10
dem3 = 1
moc = 10
while n >= moc:
    dem3 += 1
    moc *= 10
print(f"Cách 3 – Số chữ số: {dem3}")
````

## Bài 16 – Nhập vào số nguyên dương n, đếm số chữ số chẵn và chữ số lẻ

````python
n = int(input("Nhập số nguyên dương n: "))
dem_chan = 0
dem_le = 0
temp = n
while temp > 0:
    chu_so = temp % 10
    if chu_so % 2 == 0:
        dem_chan += 1
    else:
        dem_le += 1
    temp //= 10
print(f"Số chữ số chẵn: {dem_chan}")
print(f"Số chữ số lẻ: {dem_le}")
````

## Bài 17 – Nhập vào số nguyên dương n, tính tổng các chữ số của n

````python
n = int(input("Nhập số nguyên dương n: "))
tong = 0
temp = n
while temp > 0:
    tong += temp % 10
    temp //= 10
print(f"Tổng các chữ số của {n} là: {tong}")
````

## Bài 18 – Nhập vào số nguyên dương n, kiểm tra n có phải số dạng 3^k không

````python
n = int(input("Nhập số nguyên dương n: "))
temp = n
while temp > 1:
    if temp % 3 != 0:
        break
    temp //= 3
if temp == 1:
    print(f"{n} là số dạng 3^k")
else:
    print(f"{n} không phải số dạng 3^k")
````

## Bài 19 – [Fibonacci] Nhập vào số nguyên dương A, tìm số Fibonacci lớn nhất không vượt quá A

Dãy Fibonacci: 1, 1, 2, 3, 5, 8, 13, ...

````python
A = int(input("Nhập số nguyên dương A: "))
a, b = 1, 1
while b <= A:
    a, b = b, a + b
print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {a}")
````

## Bài 20 – Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố

````python
def la_nguyen_to(x):
    if x < 2:
        return False
    for i in range(2, x):
        if x % i == 0:
            return False
    return True

n = int(input("Nhập số nguyên dương n: "))
dem = 0
temp = n
while temp > 0:
    chu_so = temp % 10
    if la_nguyen_to(chu_so):
        dem += 1
    temp //= 10
print(f"Số chữ số nguyên tố trong {n}: {dem}")
````
